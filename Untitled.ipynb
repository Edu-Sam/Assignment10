```python
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt1
import plotly.graph_objects as px 
#import plotly.express as px 
import seaborn as sns
import numpy as np    
import sklearn   
from sklearn.model_selection import train_test_split
from scipy import stats
from sklearn.cluster import KMeans
```

```python
kag=pd.read_csv('KAG_conversion_data.csv')
kag
```

```python
kag.dtypes
#heart_continous=heart.select_dtypes(["float64"]).columns
```

```python
kag_categorical=kag.select_dtypes(["object"]).columns
kag[kag_categorical]
```

```python
X= pd.get_dummies(kag, columns = kag_categorical)
X
```

```python
plt1.figure(figsize=(12,10))
cor = X.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt1.show()
```

```python
X.drop(X.iloc[:, 0:4], inplace = True, axis = 1) 
X
#We dropped this 4 columns since their only use in our data frame is identification of the rows rather than
#give any useful data
```

```python
wcss = []
for i in range(2, 20):
    km = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    km.fit(X)
    wcss.append(km.inertia_)
    
plt1.plot(range(2, 20), wcss)
plt1.title('The Elbow Method', fontsize = 20)
plt1.xlabel('No. of Clusters')
plt1.ylabel('wcss')
plt1.show()

#number of appropriate clusters=5
```

```python
km = KMeans(n_clusters=5)
```

```python
xfeatures = X
clusters = km.fit_predict(xfeatures)
```

```python
y_means=km.labels_
y_means
```

```python
kag['Advert_Type']=y_means
kag
```

```python
sns.barplot(x = 'Advert_Type',
            y = 'Spent',
            hue = 'age',
            data = kag)
 
# Show the plot
plt1.show()
#According to our chart,the advert type 1 leads to the highest and consistent amount of sales.
```
